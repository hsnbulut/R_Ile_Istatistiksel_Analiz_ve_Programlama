### **15.1.5. Tek Örnek Konum Parametresi İçin R Fonksiyonu**

```r
OneSampleMeans<-function(x,
var=NULL,
H1="two.sided",
MU=0){
library(stats); library(BSDA)
if(shapiro.test(x)$p<0.05){
result<-wilcox.test(x = x,alternative = H1,
mu = MU)
}else if(missing(var)==FALSE){
result<-z.test(x = x, sigma.x=sqrt(var),
alternative = H1,mu = MU)
}else {
  result<-t.test(x = x,alternative = H1,mu = MU)
}
return(result)
}
```
