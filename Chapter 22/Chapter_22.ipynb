# **22. İLERİ DÜZEY GRAFİKLER: `ggplot2` PAKETİ**

```r
install.packages("ggplot2")
```

```r
library(ggplot2)
```

```r
iris2<-dplyr::mutate(.data = iris,Uzunluk=ifelse(Sepal.Length>5.5,"Uzun","Kısa"))
iris2$Uzunluk<-as.factor(iris2$Uzunluk)
str(iris2)
```

## **22.1. Çubuk (Sütun, Bar) Grafiği**

```r
ggplot2::qplot(data = iris2,x = Uzunluk,geom = "bar")
```

```r
ggplot2::ggplot(data = iris2,aes(x=Uzunluk))+
  geom_bar()
```

```r
ggplot(data = iris2,aes(x=Uzunluk,fill=Uzunluk))+
  geom_bar()
```

```r
ggplot(data = iris2,aes(x=Uzunluk,fill=Species))+
 geom_bar()
```

```r
ggplot(data = iris2,aes(x=Uzunluk))+
  geom_bar(fill=c("blue","red"))
```

## **22.2. Pasta (Pie) Grafiği**

```r
veri<-data.frame(Eğitim=c("İlkokul","Ortaokul","Lise","Üniversite"),frekans=c(30,50,120,150))
veri
```

```r
pp<-ggplot(data=veri,aes(x="",y=frekans,fill=Eğitim))+
 geom_bar(stat = "identity")
pp
```

```r
pp+coord_polar(theta="y",start = 0)
```

### **22.3. Histogram Grafikleri**

```r
qplot(data=iris2,x = Sepal.Length,geom = "histogram",bins=20)
```

```r
hist<-ggplot2::ggplot(data=iris2,aes(x=Sepal.Length))
hist+geom_histogram(bins =20)
```

```r
hist+geom_histogram(fill="darkred",bins=20,col="black",size=0.5)
```

```r
ggplot(data=iris2,aes(Sepal.Length,fill=Species))+
 geom_histogram(bins=20,col="black",size=0.5)
```

## **22.5. Kutu (Boxplot) Grafikleri**

```r
qplot(data=iris2,y=Sepal.Length,geom="boxplot")
```

```r
ggplot(data=iris2,aes(y=Sepal.Length))+
 geom_boxplot()
```

```r
qplot(data=iris2,x=Species,y=Sepal.Length,geom="boxplot",fill=Species)
```

```r
ggplot(data=iris2,aes(x=Species,y=Sepal.Length, fill=Species))+
 geom_boxplot()
```

## **22.7. Saçılım Grafikleri**

```r
qplot(data=iris2,x=Sepal.Length,y=Sepal.Width)
```

```r
a<-ggplot(data=iris2,aes(x=Sepal.Length,y=Sepal.Width))
a+geom_point()
```

```r
a+geom_point(color="steelblue",fill="red",size=2,shape=24)
```

```r
a+geom_point(aes(fill=Species),shape=21)
```

```r
a+geom_point(aes(fill=Species,size=Sepal.Width),shape=21)
```

## **22.8. Farklı Tür Grafiklerin Birlikte Kullanımı**

```r
x<-ggplot(data=iris2,aes(y=Sepal.Length,x=Species,fill=Species))
```

```r
x2<-x+geom_boxplot()
x2
```

```r
x2+geom_point()
```

```r
x2+geom_jitter()
```

```r
ggplot(data=iris2, aes(x = Sepal.Length))+
  geom_histogram(aes(y = ..density..),fill="white",color="black",bins=30)+
  geom_density(fill = "steelblue", alpha = 0.40)
```

```r
ggplot(data = iris,aes(Sepal.Length,Petal.Length))+
 geom_point()+
 geom_line(data=tail(iris,10),color="red",cex=2)
```
