## **10.3. Histogram**

```r
set.seed(2021)
## Rastgle not üret
not<-50+10*rnorm(100)
summary(not)
```

```r
histogram<-hist(not,col="red",xlab="Notlar",
main="Notlara Ait Histogram Grafiği")
```

```r
histogram<-hist(not,col="red",xlab="Notlar",
main="Notlara Ait Histogram Grafiği")
xfit<-seq(min(not),max(not),length=100)
yfit<-dnorm(xfit,mean=mean(not),sd=sd(not))
yfit <- yfit*diff(histogram$mids[1:2])*length(not)
lines(xfit, yfit, col="blue", lwd=2)
box()
```

## **10.7. Diğer Grafik Fonksiyonları**

```r
plot(pressure)
grid(lty = 1,lwd = 0.5,col="red")
```

```r
plot(pressure, type="b",main="Isı ve Basınç Grafiği",xlab="Isı",ylab="Basınç")
points(pressure[pressure$temperature>=175,],col="red",pch=15, cex=2)
```

```r
set.seed(2021)
X<-rnorm(100) # Standart normal dağılımdan 100 gözlem
Y<-3+2*X^2 # X'e bağlı Y değerleri
Z<-3*X^3
plot(X,Y,col="red",pch=15, xlim=c(-3,3),ylim=c(-40,40))
points(X,Z,col="blue",pch=19)
legend("topleft",c("Y","Z"),col=c("red","blue"),pch=c(15,19))
```

```r
plot(pressure, type="b",main="Isı ve Basınç Grafiği",xlab="Isı",ylab="Basınç")
lines(pressure[pressure$temperature>=175,],col="red",pch=15,lwd=3)
```

```r
plot(pressure)
abline(h=600, v=250,col=c("red","blue"),lwd=c(2,3),lty=c(1,3))
text(pressure,labels=c(1:18,"outlier"),pos=2,col="blue")
```
