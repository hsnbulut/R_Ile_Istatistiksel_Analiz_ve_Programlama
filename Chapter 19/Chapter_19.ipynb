## **19.1. t-Testleri İçin Power Analiz**

```r
library(pwr)
pwr.t.test(d=0.6325,power=0.9)
```

```r
pw<-pwr.t.test(d=0.5,power=0.9)
plot(pw)
```

## **19.2. ANOVA İçin Power Analiz**

```r
pwr.anova.test(k = 4, f=0.25,sig.level = 0.05,power=0.9)
```

```r
PowerAnaliz<-pwr.anova.test(k = 4, f=0.25,sig.level = 0.05,power=0.9)
plot(PowerAnaliz)
```
