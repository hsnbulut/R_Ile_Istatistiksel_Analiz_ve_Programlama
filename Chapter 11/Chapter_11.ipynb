## **Merkezi Limit Teoremi**

```r
set.seed(100)
x = rnorm(n=100, mean=0, sd=1)
hist(x,main="Standart Normal Dağılımdan Örneğin Histogram Grafiği")
```

```r
M=100000
ortalamalar<-NULL
for (i in 1:M){
x=rnorm(n=100,mean=0,sd=1)
ortalamalar[i]<-mean(x)
}
hist(ortalamalar,main="Ortalama İstatistiğinin Örnekleme Dağılımı (NORMAL DAĞILIM)")
```

```r
set.seed(100)
x = rexp(n=100, rate=1)
hist(x,main="Üstel Dağılımdan Örneğin Histogram Grafiği")
```

```r
M=100000
ortalamalar<-NULL
for (i in 1:M){
x = rexp(n=100, rate=1)
ortalamalar[i]<-mean(x)
}
hist(ortalamalar, main="Ortalama İstatistiğinin Örnekleme Dağılımı (ÜSTEL DAĞILIM)")
```
