## **13.1. Kovaryans**

### **`kovaryans()` fonksiyonu**

```r
kovaryans<-function(x,y,popul=FALSE){
n<-length(x)
Xort<-mean(x) ; Yort<-mean(y)
T<-sum(x*y)-n*Xort*Yort
if(popul==FALSE){
return(T/(n-1))
}else if (popul==TRUE){
return(T/n)
}else{
warning("Lütfen popul argümanına TRUE ya da FALSE giriniz!")
}}
```

```r
x<-c(30,30,25,20,18,16,12,9)
y<-c(80,70,70,65,60,50,50,35)
kovaryans(x,y)
```
