## **Örnek 3.2 Uygulama**

Üründen kaç adet almak istediğinizi aşağıdaki gibi x nesnesine giriniz. 

```r
x=20
```

Satın alınacak ürün adedine göre ödenecek tutarı belirlemek için aşağıdaki kodu çalıştırınız. Farklı ürün adetleri için farklı tutarlar belirleyebilirsiniz.


```r
if(x<10){
  tutar=10*x
} else if (x>=10 & x<50){
    tutar=8*x
    } else if (x>=50 & x<100){
    tutar=6*x
  } else {
    tutar=5*x
}
tutar
```
